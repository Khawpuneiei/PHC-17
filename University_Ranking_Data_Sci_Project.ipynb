# TOP100 SCORE(MEAN) University 2012-2015  Analysis 

```python
x = np.array([2012,2013,2014,2015])
y = np.array([])
yr=2011

for i in range(4):
  ls = []
  yr+=1
  dfyr = df[(df['year']==yr)]
  dfT100yr = dfyr[(df['world_rank']<=100)]
  for j in dfT100yr['score']:
    ls.append(j)
  Mean = statistics.mean(ls)
  y = np.append(y , [[Mean]])


plt.figure(figsize=(3,3))
plt.plot(x,y, linestyle='dashed', color='blue')
plt.title('TOP100 Score(Mean) in university ranking 2012-2015')
plt.grid()
plt.show()
```

# COUNTRY IN University Ranking 2012-2015

## COUNTRY IN TOP100 University Ranking 2012-2015

### COUNTRY IN TOP100 University Ranking 2012

```python
dfyr = df[(df['year']==2012)]
dic = {}
dicplt = {'other' : 0}
row = 0
total = 0
x = set(dfyr['country'])
y = list(x)
for i in y:
  dic.update({i:0})
for j in dfyr['country']:
  for k in dic.keys():
    if(j==k):
      total=dic[k]+1
      dic.update({k:total})
      total=0
for l in dic.keys():
  if(dic[l]>=5):
    dicplt.update({l:dic[l]})
  else:
    total = total+dic[l]
    dicplt.update({'other':total})
  row+=1

labels = dicplt.keys()
sizes = dicplt.values()
plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels,autopct='%.0f%%', startangle=90)
plt.title('country in TOP100 university ranking 2012')
plt.show()
```
